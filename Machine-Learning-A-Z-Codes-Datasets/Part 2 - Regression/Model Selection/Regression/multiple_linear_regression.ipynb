# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv(r'C:\Users\weeks\Desktop\GitHub\ML\ML\Machine-Learning-A-Z-Codes-Datasets\Part 2 - Regression\Custom Challenge\regression_model_selection_challenge.csv')
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1].values
```

Encode the neighborhood column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
transformer = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(drop="first", sparse_output=False), ["neighborhood"])
    ],
    remainder="passthrough"
)
X = transformer.fit_transform(X)
```

Skip label encoding for the neighborhood column

```python
# Neighborhood has already been one-hot encoded, so label encoding is not needed.
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
y_pred = y_pred.reshape(len(y_pred), 1)
y_test = y_test.reshape(len(y_test), 1)

np.set_printoptions(precision=2)
#print(np.concatenate((y_pred, y_test), 1))
```

## Evaluating the Model Performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
